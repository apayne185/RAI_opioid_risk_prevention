### 2.1 Basic Cleaning and Schema Alignment

```python
# Outcome prevalence and missingness
TARGET = "survived"  # target: 1 = survived, 0 = died

# Create a copy of the raw DataFrame
df = df_raw.copy()

# Standardize column names
df.columns = [c.strip().replace(" ", "_").replace("-", "_")
              for c in df.columns]

# Drop ID column
if df.shape[1] > 0:
    df = df.drop(columns=[df.columns[0]])


# Helper to coerce various binary encodings to 0/1
def to_binary(s: pd.Series) -> pd.Series:
    if s.dtype == "O":
        mapped = s.astype(str).str.strip().str.lower().map({
            "1": 1, "0": 0,
            "y": 1, "n": 0,
            "yes": 1, "no": 0,
            "true": 1, "false": 0,
            "male":1, "female":0
        })
        s = pd.to_numeric(mapped, errors="coerce")
    else:
        s = pd.to_numeric(s, errors="coerce")
    return (s.fillna(0) > 0).astype(int)


def to_date(dt: pd.Series) -> pd.Series: 
    return pd.to_datetime(dt, errors='coerce')
   
#ensure conversion to binary 
binary_cols = ['gender', 'family_history', 'hypertension', 'asthma', 'cirrhosis', 'other_cancer']
for col in binary_cols:
    df[col] = to_binary(df[col])
    print(f"{col} done")

#conversion to datetime
df["diagnosis_date"] = to_date(df["diagnosis_date"])
df["end_treatment_date"] = to_date(df["end_treatment_date"])

#one hot encoding for necessary columns
cat_cols = ['smoking_status', 'treatment_type']      #categorical columns
df = pd.get_dummies(df, columns=cat_cols, drop_first=False)

#ordinal encoding for cancer_stage
stage_map = {
    "Stage I": 1,
    "Stage II": 2,
    "Stage III": 3,
    "Stage IV": 4
}
df["cancer_stage"] = df["cancer_stage"].map(stage_map)


#for better readability
rename_dict = {
    'smoking_status_Current Smoker': 'smoker_current',
    'smoking_status_Former Smoker': 'smoker_former',
    'smoking_status_Never Smoked': 'smoker_never',
    'smoking_status_Passive Smoker': 'smoker_passive',
    'treatment_type_Chemotherapy': 'treat_chemo',
    'treatment_type_Combined': 'treat_combined',
    'treatment_type_Radiation': 'treat_radiation',
    'treatment_type_Surgery': 'treat_surgery'
}
df.rename(columns=rename_dict, inplace=True)

```

**Here we are asking if survivors have different average values than the non suvivors.**

To analyze any influence from the continuous variables on the target feature, we will be

**Survival vs Age**

There appears to be minimal relationship between age and survival rates, as the means are both ~55.0 and the both of the medians are 55.0, along with the quartiles (middle 50%) both being 42-62 years, which aligns with lung cancer being more commonafter 45. The one difference is the non-survivor gropu had patients as young as 4 and as old as 104, while the survivor group was within the range of 10-101 years. 

In the statistical tests below, the mean age differs by around ~0.03 years (essentially no difference), the t-test p=0.244 and mann-whitney u test p= 0.141 (both >0.05) show no statistically different difference, and the cohen's d= 0.003 shows a negligable effective size.
```python
plt.figure(figsize=(4,2))
sns.boxplot(x='survived', y='age', data=df);
plt.show()
```

```python
group0 = df[df["survived"] == 0]['age'].dropna()
group1 = df[df["survived"] == 1]['age'].dropna()


def cont_tests(group0, group1):     #this will be reused for all continuous vars     
    results = []
    #assumes equal variance, looking for p<0.05
    t_stat, t_p = ttest_ind(group0, group1, equal_var=True)
    #mannwhitney test, tests distribution shifts, p<0.05 
    u_stat, u_p = mannwhitneyu(group0,group1, alternative="two-sided")   

    #cohens d, tests for effect size for mean diff (0.2=small, 0.5=meduim, 0.8=large)
    mean_diff = group1.mean()-group0.mean()
    pooled_std = np.sqrt(((len(group0)-1)*group0.std()**2 + (len(group1)-1)*group1.std()**2) /(len(group0)+len(group1)-2))
    cohen_d = mean_diff / pooled_std  

    
    results.append({
        "variable": col,
        "group0_mean": group0.mean(), "group1_mean": group1.mean(),
        "t-test_p": t_p,
        "mw_p": u_p,       #mann whitney
        "cohen_d": cohen_d
    })

    return results

bivar_age_results = pd.DataFrame(cont_tests(group0, group1))
print(bivar_age_results)

```

```python
# Univariate p-values for all features
from scipy.stats import chi2_contingency, mannwhitneyu

# Identify feature types
all_cols = [c for c in df.columns if c != "survived"]
# Consider numeric as continuous; uint8/int64 binaries treated separately when 0/1
numeric_cols = [c for c in all_cols if pd.api.types.is_numeric_dtype(df[c])]
# Heuristic: binary if only {0,1}
binary_like_cols = [c for c in numeric_cols if set(df[c].dropna().unique()).issubset({0,1})]
cont_cols = [c for c in numeric_cols if c not in binary_like_cols]

uni_rows = []
# Continuous: Mann–Whitney U
for c in cont_cols:
    g0 = df.loc[df["survived"] == 0, c].dropna()
    g1 = df.loc[df["survived"] == 1, c].dropna()
    if len(g0) and len(g1):
        _, p = mannwhitneyu(g0, g1, alternative="two-sided")
        uni_rows.append({"feature": c, "test": "mannwhitney", "p_value": p})

# Binary-like: Chi-squared
for c in binary_like_cols:
    ct = pd.crosstab(df[c], df["survived"]) 
    if ct.shape == (2, 2) or ct.shape[0] >= 2:
        _, p, _, _ = chi2_contingency(ct)
        uni_rows.append({"feature": c, "test": "chi2", "p_value": p})

univariate_all = pd.DataFrame(uni_rows).sort_values("p_value")
print("Univariate p-values (lowest first):")
print(univariate_all.head(40))
```

```python
# Rebuild engineered features after dataset reset to ensure availability for modeling
# 1) Region from country, 2) treatment_duration_days from dates, 3) interaction terms

# Region mapping
region_map = {
    "Sweden": "North", "Denmark": "North", "Finland": "North", "Ireland": "North",
    "Netherlands": "West", "Belgium": "West", "Luxembourg": "West", "France": "West", "Germany": "West", "Austria": "West",
    "Italy": "South", "Spain": "South", "Portugal": "South", "Greece": "South", "Malta": "South", "Cyprus": "South",
    "Poland": "East", "Czech Republic": "East", "Slovakia": "East", "Hungary": "East", "Slovenia": "East",
    "Croatia": "East", "Romania": "East", "Bulgaria": "East", "Estonia": "East", "Latvia": "East", "Lithuania": "East"
}
# Reattach region string and one-hot encode
if "region" not in df.columns:
    df["region"] = df_raw.loc[df.index, "country"].map(region_map).fillna("Other")
region_cols_pre = [c for c in df.columns if c.startswith("region_")]
if not region_cols_pre:
    df = pd.get_dummies(df, columns=["region"], drop_first=False)

# Treatment duration from raw dates aligned to sample
if "treatment_duration_days" not in df.columns:
    _diag = to_date(df_raw.loc[df.index, "diagnosis_date"])
    _end = to_date(df_raw.loc[df.index, "end_treatment_date"])
    df["treatment_duration_days"] = (_end - _diag).dt.days

# Interaction terms with duration
if "duration_stage" not in df.columns and "treatment_duration_days" in df.columns and "cancer_stage" in df.columns:
    df["duration_stage"] = df["treatment_duration_days"] * df["cancer_stage"]
for t in ["treat_chemo", "treat_surgery", "treat_combined", "treat_radiation"]:
    if t in df.columns and f"{t}_duration" not in df.columns:
        df[f"{t}_duration"] = df[t] * df["treatment_duration_days"]
if "duration_comorbid" not in df.columns and "comorbidities" in df.columns:
    df["duration_comorbid"] = df["treatment_duration_days"] * df["comorbidities"]

# Recreate earlier interaction terms if missing
for t in ["treat_chemo", "treat_surgery", "treat_combined", "treat_radiation"]:
    if t in df.columns and f"age_{t.split('_')[1]}" not in df.columns:
        df[f"age_{t.split('_')[1]}"] = df["age"] * df[t]
    if t in df.columns and f"{t}_stage" not in df.columns:
        df[f"{t}_stage"] = df[t] * df["cancer_stage"]
    if t in df.columns and f"{t}_comorbid" not in df.columns and "comorbidities" in df.columns:
        df[f"{t}_comorbid"] = df[t] * df["comorbidities"]
    if t in df.columns and f"{t}_smoked" not in df.columns and "smoked" in df.columns:
        df[f"{t}_smoked"] = df[t] * df["smoked"]
if "age_stage" not in df.columns:
    df["age_stage"] = df["age"] * df["cancer_stage"]
if "age_comorbid" not in df.columns and "comorbidities" in df.columns:
    df["age_comorbid"] = df["age"] * df["comorbidities"]

print("Engineered features ensured. Present samples:")
print([c for c in df.columns if c.startswith("region_")])
print([c for c in df.columns if c.endswith("_duration")][:6])
print([c for c in ["duration_stage", "duration_comorbid"] if c in df.columns])
```

## 8. Feature Selection

Based on significance testing, we can now systematically select the most relevant features for modeling. This section combines statistical significance, model-based importance, and domain knowledge to create an optimal feature set.


```python
# Feature Selection: Combine statistical significance, model importance, and domain knowledge

# 1. Statistical significance-based selection
# Get top features by multivariate p-value (most significant)
significant_features = glm_all_p[glm_all_p['p_value'] < 0.05]['feature'].tolist()
print(f"Statistically significant features (p < 0.05): {len(significant_features)}")
print(significant_features[:10])

# 2. Model-based feature importance (from Lasso coefficients)
# Get absolute coefficient values as importance measure
importance_df = pd.DataFrame({
    'feature': coef.index,
    'abs_coef': np.abs(coef.values),
    'coef': coef.values
}).sort_values('abs_coef', ascending=False)

important_features = importance_df[importance_df['abs_coef'] > 0.001]['feature'].tolist()
print(f"\\nFeatures with |coef| > 0.001: {len(important_features)}")
print(important_features[:10])

# 3. Domain knowledge selection (core clinical features)
domain_features = [
    # Core predictors
    'age', 'gender', 'cancer_stage',
    # Treatment-related
    'treat_chemo', 'treat_radiation', 'treat_surgery', 'treat_combined',
    # Risk factors
    'family_history', 'smoked', 'comorbidities',
    # New engineered features that showed promise
    'treatment_duration_days', 'duration_stage', 'duration_comorbid',
    # Region (geographic variation)
    'region_East', 'region_North', 'region_South', 'region_West'
]

# 4. Conservative selection: intersection of statistical, model-based, and domain knowledge
conservative_features = list(set(significant_features) & set(important_features) & set(domain_features))
print(f"\\nConservative selection (intersection): {len(conservative_features)}")
print(conservative_features)

# 5. Expanded selection: union of statistical and model-based, filtered by domain knowledge
expanded_features = list((set(significant_features) | set(important_features)) & set(domain_features + [c for c in df.columns if c.startswith(('duration_', 'treat_', 'region_', 'age_'))]))
print(f"\\nExpanded selection (union filtered by domain): {len(expanded_features)}")

# Remove target and ID-like columns if any
final_features = [f for f in expanded_features if f != 'survived' and not f.startswith('id')]

print(f"\\nFinal feature set: {len(final_features)} features")
print("Top 20 features by model importance:")
print(importance_df[importance_df['feature'].isin(final_features)].head(20)[['feature', 'abs_coef', 'coef']])
```

## 9. Model Validation and Performance

Let's validate our feature selection and model performance using cross-validation and comprehensive metrics suitable for imbalanced classification.


```python
# Model validation with cross-validation for different feature sets

from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, average_precision_score

# Define different feature sets to compare
feature_sets = {
    'baseline': ['age', 'gender', 'cancer_stage', 'family_history', 'comorbidities', 'smoked'],
    'with_treatment': ['age', 'gender', 'cancer_stage', 'family_history', 'comorbidities', 'smoked',
                      'treat_chemo', 'treat_radiation', 'treat_surgery', 'treat_combined'],
    'with_duration': ['age', 'gender', 'cancer_stage', 'family_history', 'comorbidities', 'smoked',
                     'treat_chemo', 'treat_radiation', 'treat_surgery', 'treat_combined',
                     'treatment_duration_days'],
    'with_interactions': ['age', 'gender', 'cancer_stage', 'family_history', 'comorbidities', 'smoked',
                         'treat_chemo', 'treat_radiation', 'treat_surgery', 'treat_combined',
                         'treatment_duration_days', 'duration_stage', 'duration_comorbid'],
    'with_region': ['age', 'gender', 'cancer_stage', 'family_history', 'comorbidities', 'smoked',
                   'treat_chemo', 'treat_radiation', 'treat_surgery', 'treat_combined',
                   'treatment_duration_days', 'duration_stage', 'duration_comorbid',
                   'region_East', 'region_North', 'region_South', 'region_West'],
    'final_selected': final_features
}

# Cross-validation setup
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)

results = {}
for name, features in feature_sets.items():
    # Ensure all features exist
    available_features = [f for f in features if f in X.columns]
    if len(available_features) < len(features):
        print(f"Warning: {len(features) - len(available_features)} features missing for {name}")
        continue

    X_subset = X[available_features]

    # Update preprocessing for this feature set
    preprocess_subset = ColumnTransformer(
        transformers=[
            ("num", numeric_transformer, [c for c in available_features if c in num_cols]),
            ("bin", binary_transformer, [c for c in available_features if c in cat_like_binary_cols])
        ],
        remainder="passthrough",
        verbose_feature_names_out=False
    )

    pipe_subset = Pipeline(steps=[
        ("preprocess", preprocess_subset),
        ("lasso", LogisticRegression(penalty="l1", solver="liblinear", C=1, class_weight="balanced", random_state=RANDOM_STATE))
    ])

    # Cross-validation scores
    try:
        scores = cross_val_score(pipe_subset, X_subset, y, cv=cv, scoring='roc_auc')
        results[name] = {
            'mean_auc': scores.mean(),
            'std_auc': scores.std(),
            'n_features': len(available_features)
        }
        print(f"{name}: AUC = {scores.mean():.3f} ± {scores.std():.3f} (n_features: {len(available_features)})")
    except Exception as e:
        print(f"Error with {name}: {e}")

# Compare feature sets
results_df = pd.DataFrame(results).T.sort_values('mean_auc', ascending=False)
print("\\nFeature Set Comparison:")
print(results_df)

# Final model evaluation on test set
best_feature_set = results_df.index[0]
best_features = feature_sets[best_feature_set]
X_best = X[[f for f in best_features if f in X.columns]]

# Fit on full training data and evaluate on test
preprocess_final = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, [c for c in best_features if c in num_cols]),
        ("bin", binary_transformer, [c for c in best_features if c in cat_like_binary_cols])
    ],
    remainder="passthrough",
    verbose_feature_names_out=False
)

pipe_final = Pipeline(steps=[
    ("preprocess", preprocess_final),
    ("lasso", LogisticRegression(penalty="l1", solver="liblinear", C=1, class_weight="balanced", random_state=RANDOM_STATE))
])

pipe_final.fit(X_train[[f for f in best_features if f in X_train.columns]], y_train)

# Test set evaluation
y_pred_proba = pipe_final.predict_proba(X_test[[f for f in best_features if f in X_test.columns]])[:, 1]
y_pred = pipe_final.predict(X_test[[f for f in best_features if f in X_test.columns]])

print(f"\\nFinal Model Performance ({best_feature_set}):")
print(f"Test AUC: {roc_auc_score(y_test, y_pred_proba):.3f}")
print(f"Test Average Precision: {average_precision_score(y_test, y_pred_proba):.3f}")
print("\\nClassification Report:")
print(classification_report(y_test, y_pred))
```
